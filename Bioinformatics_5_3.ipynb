# Data visualization 

## Here we implement a k means clustering on a 2 dimensional data set and then visualize it using matplotlib library.

```python
from collections import defaultdict
def llyod_algo(data,k,n):
    centers=data[0:k]
    #print(centers)
    dict_={}
    while (n>0):
        for i in range(len(data)):
            dict_[i]=close_cent(data[i],centers)
        cluster=defaultdict(list)
        for key,val in sorted(dict_.items()):
            cluster[val].append(key)
        #print(cluster)
        for j in cluster:
            cg=cog(cluster[j],data)
            centers[j]=cg
        n=n-1
    return(centers,cluster)
```

```python
def close_cent(data,centers):
    length=0
    length_f=10000000000000
    x=0
    for i in range(len(centers)):
        length=0
        for j in range(len(centers[i])):
            length+=(data[j]-centers[i][j])**2
        if length<length_f:
            x=i
            length_f=length
    return(x)
```

```python
def cog(list1,data):
    cog_=[]
    for i in range(len(data[0])):
        s=0
        for j in (list1):
            s+=float(data[j][i])
        cog_.append(s/len(list1))    
    return(cog_)
```
